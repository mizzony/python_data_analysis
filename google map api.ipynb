```python
df_inputs = pd.read_csv('Clinic_Chiangmai.csv')
df_inputs
```

```python
data = []
for index, row in df_inputs.iterrows():
    place = getPlaceId(row[0])
    place['input'] = row[0]
    data.append(place)
```

```python
df_place_ids = pd.DataFrame(data, columns = ['input', 'status', 'place_id']) 
df_place_ids
```

```python
place_detail = []
for index, row in df_place_ids.iterrows():
    try:
        if row['status'] == 'OK':
            detail = {}
            result = getDetailByPlaceId(row['place_id'])
            result = result['result']
            detail['name'] = result.get('name')
            detail['address'] = result.get('formatted_address')
            detail['phone_number'] = result.get('formatted_phone_number')
            detail['lat'] = result.get('geometry').get('location').get('lat')
            detail['lng'] = result.get('geometry').get('location').get('lng')
            detail['website'] = result.get('website')
            place_detail.append(detail)
    except:
        print(result)
        print('--------------') 
```
